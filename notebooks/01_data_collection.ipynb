```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ["CL", "GC","NVDA","SPY","TSLA"]
```

```python
prices = yf.download(
    tickers,
    start="2015-01-01",
    end="2025-01-01",
    auto_adjust=True
)["Close"]

prices.head()
```

```python
prices.to_csv("../data/raw/prices.csv")
```

```python
prices.plot(title="Adjusted Prices", figsize=(10,5))
```
