```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ["AAPL", "MSFT","NVDA","TSLA","SPY","QQQ","TLT","BND","GC","CL","KO","PEP","BA","JPM","JNJ"]
```

```python
prices = yf.download(
    tickers,
    start="2015-01-01",
    end="2025-12-25",
    auto_adjust=True
)["Close"]

prices.head()
```

```python
prices.to_csv("../data/raw/prices.csv")
```

```python
prices.plot(title="Adjusted Prices", figsize=(10,5))
```
