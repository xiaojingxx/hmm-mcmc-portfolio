```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ["KO", "PEP"]
```

```python
prices = yf.download(
    tickers,
    start="2015-01-01",
    end="2025-01-01",
    auto_adjust=True
)["Close"]

prices.head()
```

```python
prices.to_csv("../data/raw/ko_pep_prices.csv")
```

```python
prices.plot(title="KO vs PEP Adjusted Prices", figsize=(10,5))
```
