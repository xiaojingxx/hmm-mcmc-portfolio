```python
import pandas as pd

features = pd.read_csv(
    "../data/processed/feature_matrix.csv",
    index_col=0,
    parse_dates=True
)

features.head()
```

```python
import re

pattern = re.compile(r'^([A-Z]+)')

# Extract asset names
assets = set()
for col in features.columns:
    match = pattern.match(col)
    if match:
        assets.add(match.group(1))

assets = sorted(list(assets))
```

```python
assets
```

```python
from scripts.models.hmm_utils import fit_hmm_for_asset, add_risk_scaling
```

```python
for asset in assets:
    hmm_df, _, _ = fit_hmm_for_asset(features, asset)
    hmm_df = add_risk_scaling(hmm_df, asset)

    out_path = f"../data/processed/hmm_{asset}.csv"
    hmm_df.to_csv(out_path)

    print(f"Saved {out_path}")
```
